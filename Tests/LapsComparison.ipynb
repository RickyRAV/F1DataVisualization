```python
import fastf1 as ff1
import fastf1.plotting as ff1pl
import matplotlib.pyplot as plt
```

```python
ff1.Cache.enable_cache('cache')
```

```python
ff1.plotting.setup_mpl()
```

```python
year, event, event_type = 2023, 'Jeddah', 'R'
session = ff1.get_session(year, event, event_type)
session.load()
```

```python
driver_1, driver_2 = 'RUS', 'HAM'
```

```python
data_1 = session.laps.pick_driver(driver_1)
data_2 = session.laps.pick_driver(driver_2)
team1_color = ff1pl.team_color('MER')
team2_color = ff1pl.team_color('MER')
```

```python
plt.rcParams['figure.figsize'] = [12,5]
fig, ax = plt.subplots()
x_min = 23
x_max = 50
ax.plot(data_1['LapNumber'], data_1['LapTime'], color=team1_color, label=driver_1)
ax.plot(data_2['LapNumber'], data_2['LapTime'], color=team2_color, label=driver_2)
plt.title(f"{session.event['EventName']} {session.event.year} Race\n"
         "LAPS Time Comparison")
ax.set_xlabel("Lap Number")
ax.legend(prop={"size":15})
ax.set_ylabel("Lap Time")
ax.set_xlim(left=x_min, right=x_max)
# print(data_1['LapTime'].max())
print(data_1.min()['LapTime'])
print(data_2.min()['LapTime'])
print(data_2.min()['LapTime']-data_1.min()['LapTime'])
plt.show()
```
