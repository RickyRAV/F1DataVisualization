```python
import fastf1 as ff1
import fastf1.plotting as ff1pl
import matplotlib.pyplot as plt
```

```python
ff1.Cache.enable_cache('cache')
ff1pl.setup_mpl()
```

```python
session = ff1.get_session(2023, 'Bahrain Grand Prix', 'R')
session.load()
```

```python
lec = session.laps.pick_driver('LEC')
ver = session.laps.pick_driver('VER')
```

```python
rbr_color = ff1.plotting.team_color('RBR')
fer_color = ff1.plotting.team_color('FER')
```

```python
plt.rcParams['figure.figsize'] = [13,6]
fig, ax = plt.subplots()
ax.plot(lec['LapNumber'], lec['LapTime'], color=fer_color, label='LEC')
ax.plot(ver['LapNumber'], ver['LapTime'], color=rbr_color, label='VER')
ax.legend(prop={"size":15})
plt.title(f"{session.event['EventName']} {session.event.year} Race\n"
            "LAPS Time Comparison")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
plt.show()
```
